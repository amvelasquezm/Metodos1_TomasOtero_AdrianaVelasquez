# 2.0.3 Sucesión de Fibonnaci
**1. Encuentre los primeros 20 terminos de esta sucesión.**

**2. Graficar la sucesión de Fibonnaci.**

**3. Usando la sucesión de números de Fibonnaci, calcular el número aureo y comparar con
el valor exacto**


```python
   
def fibonacci_series(n):
    fibonacci = np.array([])
    fibonacci = np.append(fibonacci, 0)
    fibonacci = np.append(fibonacci, 1)
    
    for i in range(n-2):
        fibonacci_Term = fibonacci[i]+ fibonacci[i+1]
        fibonacci = np.append(fibonacci, fibonacci_Term)
    
    return fibonacci

def f(n):
    return fibonacci_series(n)[-1]

def phi_series(n):
    phi = np.array([])
    for i in range(2, n+2):
        phi_Term = f(i+1)/f(i)
        phi = np.append(phi, phi_Term)
        
    return phi


def phi_num(x):
    return np.full(x.shape,(1+np.sqrt(5))/2)


#Encuenra los primeros 20 terminos de fibonacci

print(fibonacci_series(20))

#Grafica la sucesión fibonacci

x = np.linspace(0,20,20)

fig = plt.figure()
a = plt.plot(x, fibonacci_series(20), label = 'Serie Fibonacci')
plt.legend()
fig.show(a)

# Grafica número aureo

fig3 = plt.figure()
c = plt.plot(x, phi_series(20), label = 'Estimación usando la serie' )
d = plt.plot(x, phi_num(x) , linestyle = 'dashed', color = 'red', label = 'Número aureo' )
plt.legend()
fig3.show()
     
```

# 2.0.8 Choques de duración finita

**1. Crear el método de clase que calcule la fuerza entre partículas en la clase Particle.
Recuerde que un cuerpo no se puede aplicar fuerza a si mismo. Cada vez que se avance
en el tiempo, debe reiniciar la fuerza y la aceleración a cero, y verificar las paredes (mirar
video en bloque-neon).**

```python
class Particle:
    
    def __init__(self,r0,v0,a0,t,m=1,radius=2,Id=0):
        
        self.dt = t[1] - t[0]
        
        self.r = r0
        self.v = v0
        self.a = a0
        
        self.R = np.zeros((len(t),len(r0)))
        self.V = np.zeros_like(self.R)
        self.A = np.zeros_like(self.R)
        
        self.m = m
        self.radius = radius
        
    def GetRadius(self):
        return self.radius
        
    def GetInstantPosition(self):
        return self.r
    
    def GetMass(self):
        return self.m
    
    def Evolution(self,i):
        
        self.SetPosition(i)
        self.SetVelocity(i)
        
        self.r += self.dt*self.v
        self.v += self.dt*self.a
        
    def SetPosition(self,i):
        self.R[i] = self.r
        
    def GetPosition(self,scale=1):
        return self.R[::scale]
    
    def SetVelocity(self,i):
        self.V[i] = self.v

    def GetVelocity(self,scale=1):
        return self.V[::scale]

    def CheckLimits(self,Limits):
        
        for i in range(2):
        
            if self.r[i] + self.radius > Limits[i][1] and self.v[i] > 0.:
                self.v[i] = -1.0*self.v[i]
            if self.r[i] - self.radius < Limits[i][0] and self.v[i] < 0.:
                self.v[i] = -1.0*self.v[i]
        
    def CheckColision(self, p2,dt,K):
        if((self.GetRadius()+p2.GetRadius())>(((self.GetInstantPosition()[0]-p2.GetInstantPosition()[0])**2)+((self.GetInstantPosition()[1]-p2.GetInstantPosition()[1])**2))**0.5):
            n0 = (self.GetInstantPosition()-p2.GetInstantPosition())
            n = n0/(n0[0]**2+n0[1]**2)**0.5
            self.v = self.v + dt*(abs(self.GetInstantPosition()-p2.GetInstantPosition())**3)*K*n/(self.GetMass())
```

```python
def RunSimulation1(t,Wall,dt,K):
    r0p1 = np.array([-15.,1.])
    v0p1 = np.array([10.,0.])
    a0p1 = np.array([0.,0.])
    r0p2 = np.array([0.,-1.5])
    v0p2 = np.array([0.,0.])
    a0p2 = np.array([0.,0.])
    
    p1 = Particle(r0p1,v0p1,a0p1,t)
    p2 = Particle(r0p2,v0p2,a0p2,t)
    
    
    Wall_ = Wall.copy()
    
    
    for it in tqdm(range(len(t)), desc='Running simulation', unit=' Steps'):
        sleep(0.0001)
        p1.Evolution(it)
        p1.CheckLimits(Wall_)
        p2.Evolution(it)
        p2.CheckLimits(Wall_)
        p1.CheckColision(p2,dt,K)
        p2.CheckColision(p1,dt,K)
        
    
    return p1,p2



# Region
Limits = np.array([[-15.,15.],[-15.,15.]])
```

```python
dt = 0.05
tmax = 10
t = np.arange(0.,tmax,dt)
K = 100
Particles = RunSimulation1(t,Limits,dt,K)
Particle1 = Particles[0]
Particle2 = Particles[1]

```

```python
scale = 1
t = t[::scale]
```

```python
fig4 = plt.figure(figsize=(5,5))
ax = fig4.add_subplot(111)

def init():
    ax.set_xlim(Limits[0][0],Limits[0][1])
    ax.set_ylim(Limits[1][0],Limits[1][1])
    
def Update(i):
    
    ax.clear()
    init()
    ax.set_title(r'$ t=%.2f \ s$' %(t[i]))
    
    
    x1 = Particle1.GetPosition(scale)[i,0]
    y1 = Particle1.GetPosition(scale)[i,1]
    vx1 = Particle1.GetVelocity(scale)[i,0]
    vy1 = Particle1.GetVelocity(scale)[i,1]
    
    x2 = Particle2.GetPosition(scale)[i,0]
    y2 = Particle2.GetPosition(scale)[i,1]
    vx2 = Particle2.GetVelocity(scale)[i,0]
    vy2 = Particle2.GetVelocity(scale)[i,1]
    
    
    circle1 = plt.Circle((x1,y1),Particle1.radius, fill=True, color='k')
    ax.add_patch(circle1)
    
    circle2 = plt.Circle((x2,y2),Particle2.radius, fill=True, color='b')
    ax.add_patch(circle2)
    
    
    ax.arrow(x1,y1,vx1,vy1,color='r',head_width=0.2,length_includes_head=True)
    ax.arrow(x2,y2,vx2,vy2,color='r',head_width=0.2,length_includes_head=True)
    
    
    
Animation = anim.FuncAnimation(fig4,Update,frames=len(t),init_func=init)
fig4.show()
```

**Ejercicio 6**

**Calcular el campo de velocidades cerca de la superficie de un cilindro de radio R = 2 cm. Para esta tarea realizar los siguientes pasos:**

* Definir una discretización en los ejes x e y, donde la regi´on es: A ∈ [−4, 4] con 25 puntos en cada eje.

* Definir la función potencial del flujo

* Calcule y guarde adecuadamente el campo de velocidades usando la definición de derivada parcial central

* Dibuje el campo de velocidades usando el método: ax.quiver(x[i],y[j],Vx[i,j],Vy[i,j]).

```python
# Definir discretización de los ejes.

n = 25
x = np.linspace(-4,4,n)
y = np.linspace(-4,4,n)
X,Y = np.meshgrid(x,y)


# Definir Función potencial del flujo

def flux(x, y, V=2, R = 2):
    
    if ((R**2)/(x**2 + y**2)<=1):
        phi = V*x*(1-((R**2)/(x**2 + y**2)))
    else:
        phi = 0 
    
    return phi

# Derivadas parciales centrales

def DerX(f,x,y,h = 0.001):
    
    d = 0.
    
    if h != 0:
        d = (f(x+h, y) - f(x-h, y))/(2*h)
        
    return d

def DerY(f,x,y,h = 0.001):
    
    d = 0.
    
    if h != 0:
        d = -((f(x, y+h) - f(x, y-h))/(2*h))
        
    return d

Vx = np.zeros((n,n))
Vy = np.zeros((n,n))

for i in range(n):
    for j in range(n):
        Vx[i,j] = DerX(flux, x[i], y[j])
        Vy[i,j] = DerY(flux, x[i], y[j])
        
# Campo de Velocidades    

fig5, ax2 =plt.subplots()

#ax.scatter(X,Y)

for i in range(n):
    for j in range(n):
        ax2.quiver(x[i],y[j],Vx[i,j],Vy[i,j], color = 'b')
        
circle = plt.Circle(( 0 , 0 ), 2, fill = False, color = 'r' )
ax2.add_artist(circle)

plt.title( 'Campo de velocidades cerca de un cilindro sólido' )
plt.xlabel('Vx')
plt.ylabel('Vy')
        
plt.show()
```

```python
def f(x):
    function = 3*x**5 + 5*x**4 - x**3
    return function

def df(x):
    derivative = 15*x**4 + 20*x**3 - 3*x**2
    return derivative

def GetNewtonRaphson(f,df,xn,itmax = 100, precision=1e-5):
    
    error = 1
    it = 0
    
    while error > precision and it <= itmax:
        
        try:
            
            xn1 = xn - f(xn)/df(xn)
            
            error = np.abs(f(xn)/df(xn))
            
            #print(error)
            
        except ZeroDivisionError:
            
            print('Division por cero')
            
        it += 1
        xn = xn1
    
    if it == itmax:
        False
    else:
        return xn

def GetAllRoots(x,tolerancia=4):
    
    Roots = np.array([])
    
    for i in x:
        
        root = GetNewtonRaphson(f,df,i)
        
        if root != False:
            
            croot = np.round(root, tolerancia)
            
            if croot not in Roots:
                Roots = np.append(Roots,croot)
                
    Roots.sort()
        
    return Roots
    
x = np.linspace(-2,2,200)
Roots = GetAllRoots(x)
print(Roots)
```

**Punto 5 de Derivación**
Vamos a usar el método de diferencia finita. Tenemos los puntos $$x_{i+2} , x_{i+1} , x_{i} , x_{i-1}, x_{i-2}$$ donde la diferencia entre 2 consecutivos es $h$ y tomamos $f(x_j) $ reemplazando $x_j $por todos los anteriores salvo $x_i$. Expandiendo por taylor obtenemos:

$$
f(x_{i+2}) = f(x) + (2h)\frac{f\prime(x)}{2!} + (2h)^2\frac{f\prime\prime(x)}{3!} +  (2h)^3\frac{f\prime\prime\prime(x)}{4!} + (2h)^4\frac{f\prime\prime\prime\prime(x)}{4!} + O(h^5) \\ $$
$$
f(x_{i+1}) = f(x) + (h)\frac{f\prime(x)}{2!} + (h)^2\frac{f\prime\prime(x)}{3!} +  (h)^3\frac{f\prime\prime\prime(x)}{4!} + (h)^4\frac{f\prime\prime\prime\prime(x)}{4!} + O(h^5) \\ $$
$$
f(x_{i-1}) = f(x) - (h)\frac{f\prime(x)}{2!} + (h)^2\frac{f\prime\prime(x)}{3!} -  (h)^3\frac{f\prime\prime\prime(x)}{4!} + (h)^4\frac{f\prime\prime\prime\prime(x)}{4!} + O(h^5) \\ $$
$$
f(x_{i-2}) = f(x) - (2h)\frac{f\prime(x)}{2!} + (2h)^2\frac{f\prime\prime(x)}{3!} -  (2h)^3\frac{f\prime\prime\prime(x)}{4!} + (2h)^4\frac{f\prime\prime\prime\prime(x)}{4!} + O(h^5) \\
$$
Ahora multiplicamos estas funciones por constantes $A,B,C,D$ e intentamos encontrar sus valores para que los términos que incluyen las derivadas de órdenes menores se cancelen, esto lleva a las relaciones
$\\ $
$$
A*2h + B*h -C*h -D*2h = 0 \implies 2A+B-C-2D =0
$$
$$
A*\frac{(2h)^2}{2!} + B*\frac{h^2}{2!} +C*\frac{h^2}{2!} +D*\frac{(2h)^2}{2!} = 0 \implies 4A+B+C+4D = 0
$$
$$
A*\frac{(2h)^3}{3!} + B*\frac{h^2}{3!} -C*\frac{h^2}{3!} -D*\frac{(2h)^3}{3!} = 0 \implies 8A+B-C-8D=0
$$
$$
A*\frac{(2h)^4}{3!} + B*\frac{h^4}{4!} +C*\frac{h^4}{4!} +D*\frac{(2h)^4}{4!} = 1 \implies 16A+B+C+16D = \frac{4!}{h^4}
$$
De acá podemos obtener los valores de las constantes como $A=\frac{1}{h^4},B=-\frac{4}{h^4},C=-\frac{4}{h^4},D=\frac{1}{h^4} $
$\\ $
Con esto tenemos la siguiente expresión: $\\ $
$Af(x_{i+2}) + Bf(x_{i+1})+Cf(x_{i-1})+Df(x_{i-2}) -(A+B+C+D)f(x) = f\prime\prime\prime\prime(x) + (32A-B+C-32D)\frac{h^5}{5!} + O(h^6) \\ $
Cuando se reemplazan las constantes por sus valores y se despeja el valor de $f\prime\prime\prime\prime(x)$ se obtiene la expresión deseada, y en particular se puede observar que el término de error es $O(\frac{h^6}{h^4}) = O(h^2)$ debido a que el término de orden $O(h^5)$ tiene coeficiente principal 0, visto al reemplazar los valores de $A,B,C,D$, entonces el siguiente término con coeficiente no $0$ es de $O(h^6)$ y al despejar el $f\prime\prime\prime\prime(x)$ este se divide por $h^4$.
